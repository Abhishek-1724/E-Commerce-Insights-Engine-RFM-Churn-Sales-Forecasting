```python
import pandas as pd

# Step 1: Load data
df = pd.read_csv('data.csv', encoding='latin-1')  # Replace 'data.csv' with your filename

# Step 2: Clean and prepare data
df.dropna(subset=['InvoiceNo', 'CustomerID'], inplace=True)
df['TotalSpend'] = df['Quantity'] * df['UnitPrice']
df['Month'] = pd.to_datetime(df['InvoiceDate']).dt.to_period('M')

# Step 3: Analysis and visualization
# Monthly revenue
df.groupby('Month')['TotalSpend'].sum().plot(kind='line', title='Monthly Revenue')

# Top products
df['Description'].value_counts().head(10).plot(kind='barh', title='Top 10 Products')
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd

# Create transaction matrix
basket = (df.groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum()
          .unstack()
          .fillna(0)
          .applymap(lambda x: 1 if x > 0 else 0))  # Convert to binary (hotel encoding)

# Generate frequent itemsets
frequent_itemsets = apriori(basket, min_support=0.05, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

rules.sort_values('lift', ascending=False).head()
```

```python
# 1. Calculate RFM metrics
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (df['InvoiceDate'].max() - x.max()).days,
    'InvoiceNo': 'nunique',
    'TotalSpend': 'sum'
}).rename(columns={
    'InvoiceDate': 'Recency',
    'InvoiceNo': 'Frequency',
    'TotalSpend': 'Monetary'
})

# 2. Apply clustering
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm)
rfm['Cluster'] = KMeans(n_clusters=4, random_state=42).fit_predict(rfm_scaled)

# 3. Visualize
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=rfm,
    x='Recency',
    y='Monetary',
    hue='Cluster',
    palette='viridis',
    s=100
)
plt.title('Customer Segments by RFM')
plt.show()
```

```python
print(rfm.columns)  # Should show: Recency, Frequency, Monetary, Cluster
```

```python
# Cluster statistics
cluster_stats = rfm.groupby('Cluster').agg({
    'Recency': ['mean', 'min', 'max'],
    'Frequency': ['mean', 'median'],
    'Monetary': ['mean', 'sum', 'count']
}).round(2)

print(cluster_stats)

# Visualize segment distributions
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for i, metric in enumerate(['Recency', 'Frequency', 'Monetary']):
    sns.boxplot(data=rfm, x='Cluster', y=metric, ax=axes[i], palette='viridis')
plt.tight_layout()
```

```python
# Create cohorts based on first purchase month
df['CohortMonth'] = df.groupby('CustomerID')['InvoiceDate'].transform('min').dt.to_period('M')
cohorts = df.groupby(['CohortMonth', 'InvoiceDate'].dt.to_period('M')).agg(
    n_customers=('CustomerID', 'nunique')
).reset_index()

# Calculate retention matrix
cohort_pivot = cohorts.pivot_table(index='CohortMonth', 
                                 columns='InvoiceDate', 
                                 values='n_customers')

# Visualize retention heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(cohort_pivot, annot=True, fmt='.0f', cmap='Blues')
plt.title('Monthly Customer Retention Rates')
plt.ylabel('Cohort Month')
plt.xlabel('Order Month')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Create churn labels (90 days inactive)
cutoff_date = df['InvoiceDate'].max() - pd.Timedelta(days=90)
churn_labels = df.groupby('CustomerID')['InvoiceDate'].max().lt(cutoff_date).astype(int)

# Merge RFM features with churn labels
model_data = rfm.merge(churn_labels.rename('Churn'), left_index=True, right_index=True)

# Train/test split
X = model_data[['Recency', 'Frequency', 'Monetary']]
y = model_data['Churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Train model
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# Evaluate
from sklearn.metrics import classification_report
print(classification_report(y_test, clf.predict(X_test)))

# Feature importance
pd.Series(clf.feature_importances_, index=X.columns).plot(kind='barh')
```
